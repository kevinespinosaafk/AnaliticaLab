```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


n = 100
X = np.linspace(0,180,n)
y = -2.0 * X + np.random.normal(0 ,25 ,n)
y_real = -2.0 * X
plt.figure(figsize=(6,4))
ax= plt.gca()

ax.scatter(X, y, alpha= 0.75)
ax.plot(X, y_real, "-")
ax.set_title("dispercion: relacion entre dos variables")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True, linestyle= ":", alpha= 0.4)

plt.show()
```
